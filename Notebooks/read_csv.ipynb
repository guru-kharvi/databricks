```python
#read csv file to dataframe
df_csv = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/dev/bronze/excel/Sales.csv")
df_csv.show()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType

# Define schema manually
manual_schema = StructType([
    StructField("OrderID", IntegerType(), True),
    StructField("Product", StringType(), True),
    StructField("Quantity", IntegerType(), True),
    StructField("Price", DoubleType(), True),
    StructField("OrderDate", StringType(), True)
])

df_csv_manual = spark.read.format("csv").option("header", True).schema(manual_schema).load("/Volumes/dev/bronze/excel/Sales.csv")
display(df_csv_manual)
```

```python
%sql
EXPLAIN EXTENDED SELECT * FROM df_csv_manual;
```

```python
# You can inspect malformed rows by adding a _corrupt_record column to your schema.
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType

schema_with_corrupt = StructType([
    StructField("OrderID", IntegerType(), True),
    StructField("Product", StringType(), True),
    StructField("Quantity", IntegerType(), True),
    StructField("Price", DoubleType(), True),
    StructField("OrderDate", StringType(), True),
    StructField("_corrupt_record", StringType(), True)
])

df_permissive_with_corrupt = spark.read.format("csv").option("header", True).option("mode", "PERMISSIVE").schema(schema_with_corrupt).load("/Volumes/dev/bronze/excel/Sales.csv")
display(df_permissive_with_corrupt)
# display(df_permissive_with_corrupt.filter(df_permissive_with_corrupt["_corrupt_record"].isNotNull()))
```

```python
# Example for PERMISSIVE mode (default)
df_permissive = spark.read.format("csv").option("header", True).option("mode", "PERMISSIVE").schema(manual_schema).load("/Volumes/dev/bronze/excel/Sales.csv")
display(df_permissive)

# Example for DROPMALFORMED mode
df_dropmalformed = spark.read.format("csv").option("header", True).option("mode", "DROPMALFORMED").schema(manual_schema).load("/Volumes/dev/bronze/excel/Sales.csv")
display(df_dropmalformed)

# Example for FAILFAST mode
df_failfast = spark.read.format("csv").option("header", True).option("mode", "FAILFAST").schema(manual_schema).load("/Volumes/dev/bronze/excel/Sales.csv")
display(df_failfast)
```
