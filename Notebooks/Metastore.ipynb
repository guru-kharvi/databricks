```python
%sql
-- remove dublicate
CREATE OR REPLACE TABLE bronze.emp AS
SELECT DISTINCT *
FROM bronze.emp;
```

CREATE OR REPLACE TABLE bronze.emp AS
SELECT empno, ename, dept_code
FROM (
    SELECT *,
           ROW_NUMBER() OVER (
               PARTITION BY empno
               ORDER BY empno
           ) AS rn
    FROM bronze.emp
)
WHERE rn = 1;
