```python
from pyspark.sql.functions import col, lit
# Create a text widget for employee ID input
dbutils.widgets.text("id", '', "emp_id")
_id = dbutils.widgets.get("id").strip() if dbutils.widgets.get("id") else None

# Read sales table from dev.bronze schema
sales_df = spark.read.table("dev.bronze.sales")

# Filter sales records by sale_id matching the input ID
filtered_sales = sales_df.filter(col("sale_id") == _id if _id is not None else lit(None))

# Display the filtered sales records
# display(filtered_sales)

# Count the number of filtered records
_count = filtered_sales.count()

# If records exist, overwrite and save them to a new table named with the input ID
if _count > 0:
    filtered_sales.write.mode("overwrite").saveAsTable(f"dev.bronze.sales_{_id}")
else:
    print(f"No date for {_id}")
dbutils.notebook.exit(_count)
```
