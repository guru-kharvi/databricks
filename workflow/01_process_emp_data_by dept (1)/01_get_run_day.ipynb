```python
from datetime import datetime

# Create a text widget for input_date
dbutils.widgets.text("input_date", "")
input_date = dbutils.widgets.get("input_date")

# Parse the input date
dt = datetime.strptime(input_date, "%Y-%m-%d %H:%M:%S")

# Python: Monday=0 ... Friday=4 ... Sunday=6
is_friday = dt.weekday() == 4

print(f"Input date: {input_date}")
print(f"Is Friday: {is_friday}")

# Store the result in task values
dbutils.jobs.taskValues.set("input_day", is_friday)
```
